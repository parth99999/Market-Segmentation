```python
import pandas as pd
import numpy as np
```

```python
data=pd.read_csv("Mall_Customers.csv")
```

```python
data_=data.iloc[:,[3,4]]
```

```python
from sklearn.cluster import KMeans
```

```python
wcss=[]
for i in range(1,11):
    k=KMeans(n_clusters=i,n_jobs=4)
    km=k.fit(data_)
    wcss.append(km.inertia_)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(range(1,11),wcss)
```

```python
k=KMeans(n_clusters=5,random_state=0)
y_kmeans=k.fit_predict(data_)
y_kmeans
```

```python
X=data_.values
X[:,0]
```

```python
plt.scatter(X[y_kmeans==0,0], X[y_kmeans==0, 1], s=50, cmap='viridis',label="Careful",c="red")
plt.scatter(X[y_kmeans==1,0], X[y_kmeans==1, 1], s=50, cmap='viridis',label="Standard",c="blue")
plt.scatter(X[y_kmeans==2,0], X[y_kmeans==2, 1], s=50, cmap='viridis',label="Target",c="green")
plt.scatter(X[y_kmeans==3,0], X[y_kmeans==3, 1], s=50, cmap='viridis',label="Careless",c="purple")
plt.scatter(X[y_kmeans==4,0], X[y_kmeans==4, 1], s=50, cmap='viridis',label="Sensible",c="yellow")
plt.scatter(k.cluster_centers_[:,0],k.cluster_centers_[:,1],c="black",s=100,label="Centroids")
plt.legend()
plt.xlabel("Annual Income")
plt.ylabel("Spending Score")

plt.show()
```

```python
data4=data.iloc[:,[1,4]]
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
```

```python
data4["Genre"]=le.fit_transform(data4["Genre"])
```

```python
X4=data4.values
wcss=[]
for i in range(1,11):
    k=KMeans(n_clusters=i,n_jobs=4)
    km=k.fit(X4)
    wcss.append(km.inertia_)
```

```python
plt.plot(range(1,11),wcss)
```

```python
k_=KMeans(n_clusters=3,random_state=0)
y_=k_.fit_predict(X4)
```

```python
plt.scatter(X4[y_==0,0], X4[y_==0, 1], s=50, cmap='viridis',label="Low Spending",c="red")
plt.scatter(X4[y_==1,0], X4[y_==1, 1], s=50, cmap='viridis',label="Standard(Promotion)",c="blue")
plt.scatter(X4[y_==2,0], X4[y_==2, 1], s=50, cmap='viridis',label="High Spending(Target)",c="green")
plt.legend()
plt.show
```
